# Khởi tạo node và element tấm

```python
# Tao một hàm tạo node và element của phần tử tấm 
def node_element(num_elements,width,lenght ):
    matrix =  np.zeros([num_elements+1,num_elements+1])
    dem = 1
    for i in range(num_elements+1):
        for j in range(num_elements+1):
            matrix[i,j] = dem
            dem +=1

    node = np.zeros((num_elements+1)**2)
    for i in range(node.shape[0]):
        node[i] = i+1
    X = np.zeros((num_elements+1)**2)
    for i in range(X.shape[0]):
        X[i]=width*(i % (num_elements+1))/num_elements
    Y = np.zeros((num_elements+1)**2)
    for i in range(Y.shape[0]):
        Y[i]=lenght*(i // (num_elements+1))/num_elements
    Z = np.zeros((num_elements+1)**2)     
    
    matrix_node = np.zeros([(num_elements+1)**2,4])
    matrix_node[:,0]=node
    matrix_node[:,1]=X
    matrix_node[:,2]=Y
    matrix_node[:,3]=Z
    
    matrix_ele = np.zeros(4)
    ele = np.zeros(4)

    for i in range(num_elements):
        for j in range(num_elements):
            ele[0] = matrix[i,j]
            ele[1] = matrix[i+1,j] 
            ele[2] = matrix[i+1,j+1] 
            ele[3] = matrix[i,j+1] 
            matrix_ele = np.vstack((matrix_ele, ele))
       
                 
    matrix_ele = np.delete(matrix_ele,0,axis=0)
    num_ele = np.zeros([num_elements**2,1])
    for i in range(num_ele.shape[0]):
        num_ele[i,:]=i+1
    matrix_ele = np.concatenate((num_ele, matrix_ele), axis=1)  
    return matrix_node,  matrix_ele
```

```python
a , b =  node_element(10,1,1)
print(b)
print(a)
```
